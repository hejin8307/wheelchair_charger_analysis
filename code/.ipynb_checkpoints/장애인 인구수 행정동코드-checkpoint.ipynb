```python
import csv
import pandas as pd
file1 = pd.read_csv("C:/Users/user/Desktop/internship/project/data/행정동 지도/행정코드(2020.6).csv", encoding='cp949')

file2 = pd.read_csv("C:/Users/user/Desktop/internship/project/data/consume/행정동별 장애인 복지카드 소비데이터(2014~2015).csv", encoding = 'utf-8')
```

```python
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 100)

left_join = pd.merge(file2, file1, left_on = '행정동 이름', right_on = '동이름', how = 'left')
left_join
```

```python
left_join.isnull().sum()
```

```python
a = left_join['dong'] == '기타'
pop = left_join[a]
pop
```

```python
remove = left_join[left_join['dong'] == '기타'].index
pop = left_join.drop(remove)
pop
```

```python
pop.isnull().sum()
```
