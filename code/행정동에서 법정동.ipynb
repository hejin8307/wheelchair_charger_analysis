```python
import csv
import pandas as pd
file1 = pd.read_excel("C:/Users/user/Desktop/internship/project/data/emd/KIKmix.20210726(말소코드포함).xlsx")

pop = pd.read_csv("C:/Users/user/Desktop/internship/project/data/pop/disabled_pop.csv", encoding = 'cp949')

is_seoul = file1['시도명'] == '서울특별시'
seoul = file1[is_seoul]
seoul
```

```python
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 100)

left_join = pd.merge(pop, seoul, left_on = 'dong', right_on = '읍면동명', how = 'left')
left_join.rename(columns={'dong':'행정동1'}, inplace = True)
left_join.rename(columns={'읍면동명':'행정동2'}, inplace = True)
left_join.rename(columns={'동리명':'법정동'}, inplace = True)
left_join
```

```python
hangjeongdong_list = left_join['행정동1'].tolist()
hangjeongdong = []
for i in hangjeongdong_list:
    hangjeongdong.append(i)
print(hangjeongdong)

# c = 0
# for i in range(0, len(left_join['행정동1'])) :
#     if left_join['행정동1'] == '사직동'
#     count += 1
```
