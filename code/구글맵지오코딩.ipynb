```python
ls = address #모르는 지역명
```

```python
result=[]
for a in ls:
    geocode_result = gmaps.geocode((a), language='ko')
    address = geocode_result[0]['formatted_address']
    x = geocode_result[0]['geometry']['location']['lat']
    y = geocode_result[0]['geometry']['location']['lng']
    result.append({ a: {'주소': address, '위도': x, '경도': y}})
print(result) #이부분은 확인용. 리스트가 길어지면 지울 것
```
